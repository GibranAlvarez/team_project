```python
import sqlite3
import pandas as pd

# Database filename
db_filename = 'Data_Science_Team_Project.db'

try:
    # Connect to the database
    conn = sqlite3.connect(db_filename)

    # Query to retrieve emigrants data
    query_returning_emigrants = "SELECT year, total_returning_emmigrants_per_year FROM total_returning_emmigrants_table WHERE year BETWEEN 1998 AND 2022;"
    returning_emigrants_data = pd.read_sql_query(query_returning_emigrants, conn)

    # Query to retrieve incidents data
    query_incidents = "SELECT year, number FROM incidents WHERE year BETWEEN 1998 AND 2022;"
    incidents_data = pd.read_sql_query(query_incidents, conn)

    # Close connection
    conn.close()
except sqlite3.Error as e:
    print(f"SQLite error: {e}")



# Merge data on 'year' column
combined_data = pd.merge(returning_emigrants_data, incidents_data, on='year', how='inner')
print(combined_data)



# Calculate Pearson correlation coefficient
correlation = combined_data['total_returning_emmigrants_per_year'].corr(combined_data['number'])
print(f"Pearson correlation coefficient: {correlation}")


import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(combined_data['total_returning_emmigrants_per_year'], combined_data['number'], alpha=0.5)
plt.title('Relationship between Returning Emigrants Count and Incidents Count')
plt.xlabel('total_returning_emmigrants_per_year')
plt.ylabel('number of incidents')
plt.grid(True)
plt.show()
```
